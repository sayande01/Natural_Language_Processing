## TF-IDF (Term Frequency-Inverse Document Frequency)

TF-IDF is a statistical measure that evaluates the importance of a word in a document relative to a collection of documents (corpus). Itâ€™s widely used in Natural Language Processing and Information Retrieval to help identify significant words and phrases in documents. Unlike simple word counts or Bag of Words, TF-IDF considers how unique or "rare" a word is across documents, assigning higher weights to less common but potentially more meaningful words.

image.png

```python
  from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text data
documents = [
    "Data science is transforming the healthcare industry by providing insights from data analysis.",
    "The finance industry is experiencing a revolution with blockchain technology providing secure transactions.",
    "Artificial intelligence and data science are advancing rapidly in the field of autonomous vehicles."
]


# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the documents to TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Convert to array and display results
tfidf_array = tfidf_matrix.toarray()
print("TF-IDF Vocabulary:", tfidf_vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_array)
```


### Interpretation of TF-IDF Results

The TF-IDF scores reveal which words are **important** within each document:

- **Document 1 (Healthcare)**: Words like â€œhealthcare,â€ â€œinsights,â€ and â€œanalysisâ€ are likely to have higher TF-IDF scores in Document 1, as they specifically highlight the healthcare topic.
- **Document 2 (Finance and Blockchain)**: Words such as â€œfinance,â€ â€œblockchain,â€ and â€œtransactionsâ€ will have higher scores here, distinguishing this document as focusing on finance.
- **Document 3 (AI and Vehicles)**: Terms like â€œartificial intelligence,â€ â€œvehicles,â€ and â€œautonomousâ€ will stand out as high-scoring in Document 3.

### Why TF-IDF Helps Capture Importance

- **Focuses on Unique Terms**: TF-IDF downplays commonly used words like "data" and "industry" because they appear in multiple documents.
- **Highlights Key Terms**: Words specific to each topic (e.g., "healthcare" in Document 1) receive higher scores, capturing their **importance** to that document.

### Applications of TF-IDF

- **Information Retrieval**: Search engines use TF-IDF to rank documents by relevance to a search query, highlighting documents with the most "important" terms.
- **Topic Identification**: TF-IDF helps extract keywords that represent the main focus of each document.
- **Text Classification**: By identifying important terms, TF-IDF aids in creating feature sets for training text classifiers.

In summary, TF-IDF is a powerful tool that allows us to understand which words best represent the core content of a document by capturing the **importance** of words relative to a collection of texts.
